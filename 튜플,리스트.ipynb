```python
# 튜플은 불변, 리스트는 가변
# chatGPT에게 질문 : 튜플과 range()함수의 차이점은 무엇?
# 답변 :
# 1. 가변성: 튜플은 한 번 생성되면 수정할 수 없지만, range() 객체는 범위를 변경할 수 있습니다.
# 2. 인덱싱과 슬라이싱: 튜플은 인덱스와 슬라이싱을 지원하고, range() 객체는 슬라이싱을 통해 새로운 객체를 생성해야 합니다.
# 3. 메모리 사용: 튜플은 요소들을 메모리에 직접 저장하며, range() 객체는 필요에 따라 숫자를 생성하여 메모리를 절약합니다.
# 4. 용도: 튜플은 데이터를 변경하지 않고 저장할 때 사용하고, range() 함수는 정수 시퀀스를 생성할 때 사용합니다.

#### 튜플
# 한 요소 이상의 튜플을 만들기 위해서는 각 요소 뒤에 콤마(,)를 붙임
one_marx = ('Gruouho',)
print(one_marx)
# 튜플을 생성할때 주의할 점 : 괄호 안에 요소만 있고, 콤마를 생략하면 튜플이 아니라 문자열이 됨
one_marx =('Groucho') # 문자열
print(one_marx)
# 요소가 두개 이상이면 마지막에는 콤마를 붙이지 않음
marx_tuple = ("1","2","3")
print(marx_tuple)
# 튜플로 한 번에 여러 변수를 할당할 수 있음
marx_tuple = ('G',"C","H")
a,b,c = marx_tuple
print(a)
print(b)
print(c)
# 한 문장에서 값을 교환하기 위해 임시변수를 사용하지 않고 튜플을 사용할 수 있음
password = 'sowrd'
icecream = 'ice'
password, icecream = icecream, password
print(password)
print(icecream)

## 생성하기 : tuple()
mlist = ['g','c','i']
tuple(mlist)

## 결합하기 : +
('G',)+('C','H') # 출력 : ('G', 'C', 'H')

## 항목 복제하기 : *
('yada',) * 3 # 출력 : ('yada', 'yada', 'yada')

## 비교하기
a = (7,2)
b = (7,2,9)
a ==b # 출력 : False
a <= b # 출력 : True

## 수정하기 : 불변이므로 연산을 하게 되면 새로운 주소값을 가진 튜플이 생성됨
a + b # 출력 : (7, 2, 7, 2, 9)

id(a) # 2656460598784
a += b
a # 출력 : (7, 2, 7, 2, 9)
id(a) # 2656460976624
```
